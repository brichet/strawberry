### 2.2.2 visualization of the most central individuals

#### 2.2.2.1 Extraction of the most central individuals

+ Extract MTG data at plant scale (function variable.extract_at_plant_scale)

```python
# Extraction of data at plant scale
df_at_plant_scale = variables.extract_at_plant_scale(g)
# Transform date as date value
df_at_plant_scale.date = pd.to_datetime(df_at_plant_scale.date, dayfirst=True)
df_at_plant_scale
```

+ Function to calculate the most central individuals 

```python
# Function to select the most central individual with all variable 
df_individuals = analysis.median_individuals(df_at_plant_scale)
df_individuals
```

+ visualization of spatio-temporal developement using 2D visualisation

```python
# selection of vid of median individuals
pids = list(df_individuals.vid)
n = len(pids)

# Plot 2D- visualisation of the most central plant
visualisation2d.plot2d(g, pids, dist=[6]*n)
```
