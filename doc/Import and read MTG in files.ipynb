## 1.1 File import and selection

Import all .mtg files in share/data files

```python
files = shared_data(openalea.strawberry).glob('*.mtg') # put in files all .mtg files present in share/data folder
print files # display path of mtg files 
```

Now that we have collected all files .mtg,  we want to select them.  
For this we create a dictionnary containing names of mtg files

```python
mtg_names = dict((f.namebase, f) for f in files) # create a dictionnary contains names of each file corresponding to MTG names
print(mtg_names.keys()) # display names of mtg files
```

Selection of the file or files you want to use with the mtg name file and store in mtg.file

```python
mtgfile = ['Gariguette','Ciflorette']
mtgfile = dict((k,f) for k,f in mtg_names.items() if k in mtgfile)
print mtgfile
```
